## Algerian Forest Fires Dataset

Data Set Information:

The dataset includes 244 instances that regroup a data of two regions of Algeria,namely the Bejaia region located in the northeast of Algeria and the Sidi Bel-abbes region located in the northwest of Algeria.

122 instances for each region.

The period from June 2012 to September 2012.
The dataset includes 11 attribues and 1 output attribue (class)
The 244 instances have been classified into fire(138 classes) and not fire (106 classes) classes.

Attribute Information:

1. Date : (DD/MM/YYYY) Day, month ('june' to 'september'), year (2012)
Weather data observations
2. Temp : temperature noon (temperature max) in Celsius degrees: 22 to 42
3. RH : Relative Humidity in %: 21 to 90
4. Ws :Wind speed in km/h: 6 to 29
5. Rain: total day in mm: 0 to 16.8
FWI Components
6. Fine Fuel Moisture Code (FFMC) index from the FWI system: 28.6 to 92.5
7. Duff Moisture Code (DMC) index from the FWI system: 1.1 to 65.9
8. Drought Code (DC) index from the FWI system: 7 to 220.4
9. Initial Spread Index (ISI) index from the FWI system: 0 to 18.5
10. Buildup Index (BUI) index from the FWI system: 1.1 to 68
11. Fire Weather Index (FWI) Index: 0 to 31.1
12. Classes: two classes, namely Fire and not Fire

```python
## Independent And dependent features
X=df.drop('FWI',axis=1)
y=df['FWI']
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train.shape,X_test.shape
```

```python
## Feature Selection based on correlaltion
X_train.corr()
```

```python
## Check for multicollinearity
plt.figure(figsize=(12,10))
corr=X_train.corr()
sns.heatmap(corr,annot=True)
```

```python
def correlation(dataset, threshold):
    # Create an empty set to store names of highly correlated columns
    col_corr = set()

    # Compute the correlation matrix for the dataset (numeric columns only)
    corr_matrix = dataset.corr()

    # Loop over all columns in the correlation matrix
    for i in range(len(corr_matrix.columns)):

        # Loop over columns before the current column
        # (to avoid duplicate comparisons and self-correlation)
        for j in range(i):

            # Check if the absolute correlation value exceeds the given threshold
            if abs(corr_matrix.iloc[i, j]) > threshold:

                # Get the name of the correlated column
                colname = corr_matrix.columns[i]

                # Add the column name to the set
                col_corr.add(colname)

    # Return the set of highly correlated column names
    return col_corr
```

```python
## threshold--Domain expertise
corr_features=correlation(X_train,0.85)
```

```python
## Drop highly correlated features to remove duplicate information
## and reduce multicollinearity, which improves model stability
X_train.drop(corr_features, axis=1, inplace=True)

## Drop the same features from test data to keep traintest feature alignment
X_test.drop(corr_features, axis=1, inplace=True)

## Check the shape after removing redundant features
X_train.shape, X_test.shape
```

# Feature Scaling Or Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
X_train_scaled
```

## Box Plots To understand Effect Of Standard Scaler

```python
plt.subplots(figsize=(15, 5))
plt.subplot(1, 2, 1)
sns.boxplot(data=X_train)
plt.title('X_train Before Scaling')
plt.subplot(1, 2, 2)
sns.boxplot(data=X_train_scaled)
plt.title('X_train After Scaling')
```

# Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
linreg=LinearRegression()
linreg.fit(X_train_scaled,y_train)
y_pred=linreg.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)

# 45 diagonal (y = x) represents perfect prediction:
# points close to this line mean predicted values  actual values
plt.scatter(y_test,y_pred)
```

# Lasso regression

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score

# Lasso regression with default alpha value
# If alpha is not specified, sklearn uses alpha = 1.0
# This value is fixed and NOT optimized for the dataset
lasso=Lasso()
lasso.fit(X_train_scaled,y_train)
y_pred=lasso.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)

# 45 diagonal (y = x) represents perfect prediction:
# points close to this line mean predicted values  actual values
plt.scatter(y_test,y_pred)
```

# Cross validation lasso
Lasso  alpha is assumed

LassoCV  alpha is learned

```python
from sklearn.linear_model import LassoCV
lassocv=LassoCV(cv=5) # Create Lasso model that automatically selects best alpha using 5-fold CV
lassocv.fit(X_train_scaled,y_train)
```

```python
lassocv.alpha_ # Best alpha value selected by cross-validation
```

```python
lassocv.alphas_ # All alpha values tested during cross-validation
```

```python
lassocv.mse_path_ # Mean squared error for each alpha across CV folds
```

```python
y_pred=lassocv.predict(X_test_scaled)
plt.scatter(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
```

# Ridge regression model

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
ridge=Ridge()
ridge.fit(X_train_scaled,y_train)
y_pred=ridge.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
plt.scatter(y_test,y_pred)
```

# Cross validation Ridge

```python
from sklearn.linear_model import RidgeCV
ridgecv=RidgeCV(cv=5)
ridgecv.fit(X_train_scaled,y_train)
y_pred=ridgecv.predict(X_test_scaled)
plt.scatter(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
```
